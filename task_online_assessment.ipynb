```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("/content/booking.csv")
data.sample(5)
```

```python
data.info()
```

```python
data2 = pd.read_csv("/content/booking_status.csv")
data2
```

```python
data3 = pd.read_csv("/content/currency.csv")
data3
```

```python
data3.loc[data3['currency_label'] == "?", 'currency_label'] = "€"
data3 #merubah simbol mata uang euro
```

```python
data4 = pd.read_csv("/content/exchange.csv")
data4.head()
```

```python
data4.info()
```

```python
df = data.merge(data2, 
                left_on = 'booking_status',
                right_on = 'booking_status',
                how = 'inner')
df.head()
```

```python
df2 = df.merge(data3, 
               left_on="booking_currency",
               right_on="currency_code",
               how="inner")
df2.sample(8)
```

```python
df2['date'] = df2['booking_received_timestamp'].str[:10]
df2.head()
```

```python
df3 = df2.merge(data4, 
                left_on='date',
                right_on="id",
                how='inner')
df3.head()
```

```python
df3.info()#data berkurang karena pada data exchange hanya dapat match terhadap 8944 observasi data
```

```python
df3.columns
```

```python
cat = ['booking_received_timestamp','booking_check_in','booking_check_out',
       'booking_status_desc','currency_label',
       'currency_desc', 'date_x', 'id', 'date_y', 'base_currency']
```

```python
for col in cat:
    print(f'''Value count kolom {col}:''')
    print(df3[col].value_counts())
    print()
```

```python
df3.groupby(['currency_desc','booking_status_desc','booking_check_out','booking_check_in']).sum()
#untuk melihat pembayaran dengan USD (CANCEL & CONFIRMED) DAN IDR(CANCEL & CONFIRMED)
```

```python
df3['amount']=np.where(df3['currency_desc']=='IDR', df3['booking_earned']/df3['idr'], df3['booking_earned'])
df3.head()
```

```python
df4 = df3.groupby(['date_x']).sum()[['amount']].reset_index()
df4.head()#date_x ini adalah pengambilan tanggal dari pemesanan dan amount adalah pendapatan per tanggal pemesanan
```

```python
df4.rename(columns={'date_x': 'date', 'amount': 'revenue'}, inplace=True)
```

```python
df4.to_csv("revenue_per_day.csv", index=False)
```
